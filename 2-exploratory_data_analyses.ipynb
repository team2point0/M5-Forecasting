```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/m5-forecasting-accuracy/combined_long_data.csv')
```

```python
df.head()
```

```python
def plotProductCount(data, by, times = 'week', kind='line', ax=None):
    ax = ax or plt.gca()
    myplot = data.groupby(times)[by].value_counts().unstack().plot(kind=kind,ax=ax)
    plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
    plt.title(by)
    plt.rcParams['figure.dpi'] = 100
    plt.figure(figsize=(6,4))
    return myplot
```

```python
plotProductCount(df, 'store_id')
```

```python
plotProductCount(df, 'state_id')
```
